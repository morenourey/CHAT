Reemplazo de valores nulos de la columna salary. Como la amplitudad de valores es muy larga, luego será necesario que se utilize una transformación por log.

```python
df_sociodemo['log_salary'] = np.log(df_sociodemo['salary'])
```

```python
#visualización de la distribución para columna

media = np.mean(df_sociodemo['log_salary'])
mediana = np.median(df_sociodemo['log_salary'])


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

sns.boxplot(x='log_salary', data=df_sociodemo, ax=axes[0])
axes[0].set_title('log_salary')


sns.histplot(x='log_salary', data=df_sociodemo, bins = 30, ax=axes[1])
axes[1].set_title('log_salary')
plt.axvline(x=media, color='red', linestyle='--', label=f'Mean: {media}')


plt.tight_layout()
plt.show()
```

```python
df_sociodemo['log_salary'].describe()
```

Para esa distribución, parece interesante reemplazar los valores nulos con el método backfill - duda, sería mejor utilizar la mediana? Con backfill, la distribución, increiblemente, está muy parecida al original

```python
 df_sociodemo['log_salary'].fillna(method='bfill', inplace = True)
```
